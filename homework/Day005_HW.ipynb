```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

'''
Your Code
'''
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

```

```python
import json
datas = json.loads(response)
datas
```

```python
len(datas)
```

```python
for data in datas:
    print(data.keys())
    print("\n")
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
comments=0
likes=0
'''
Your Code
'''
for data in datas:
    print('æ¨™é¡Œ:{}'.format(data['title']))
    print('è²¼æ–‡æ™‚é–“:{}'.format(data['createdAt']))
    print('ç•™è¨€äººæ•¸:{}'.format(data['commentCount']))
    comments=comments+data['commentCount']
    print('æŒ‰è®šæ•¸:{}'.format(data['likeCount']))
    likes=likes+data['likeCount']
    print('\n')
```
