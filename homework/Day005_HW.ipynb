```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

```

```python
import json
datas = json.loads(response)
datas
```

```python
len(datas)
```

```python
for data in datas:
    print(data.keys())
    print("\n")
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
comments=0
likes=0
'''
Your Code
'''
for data in datas:
    print('標題:{}'.format(data['title']))
    print('貼文時間:{}'.format(data['createdAt']))
    print('留言人數:{}'.format(data['commentCount']))
    comments=comments+data['commentCount']
    print('按讚數:{}'.format(data['likeCount']))
    likes=likes+data['likeCount']
    print('\n')
```

```python
print("寵物版熱門文章平均留言人數:{}".format(comments/30))
print("寵物版熱門文章平均按讚人數:{}".format(likes/30))
```
